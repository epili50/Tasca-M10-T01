El siguiento código extrae de la página de eltiempo.es la predicción tempertura y la velocidad el tiempo de Barcelona del día en que se ejecuta, hora por hora 

[Link de la página rápada](https://www.eltiempo.es/barcelona.html?v=por_hora)

```python
# Librerías
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

# Opciones de navegación
options =  webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_argument('--disable-extensions')

driver_path = (r"/Users/macbook/Documents/Personal/Progamacion/Sprints/10º Modulo/chromedriver")

driver = webdriver.Chrome(driver_path, chrome_options=options)
```

```python
driver.get('https://www.eltiempo.es/barcelona.html?v=por_hora')

#Acetpo las cookies
WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.CSS_SELECTOR,
                                      'button.didomi-components-button didomi-button didomi-dismiss-button didomi-components-button--color didomi-button-highlight highlight-button'.replace(' ', '.'))))\
    .click()
```

```python
#selecciono el cuadro con las fechas y los datos del tiempo
WebDriverWait(driver, 5)\
    .until(EC.element_to_be_clickable((By.XPATH,
                                      '/html/body/div[5]/div[1]/div[4]/div/section[4]/section/div[1]/ul')))
                                        
#Tomo los elementos del cuadro
texto_columns = driver.find_element("xpath",'/html/body/div[5]/div[1]/div[4]/div/section[4]/section/div[1]/ul')

#covierto los elementos en texto
texto_columns = texto_columns.text

#Recorto lo datos de hoy
tiempo_hoy = texto_columns.split('Mañana')[0].split('\n')[1:-2]


print(tiempo_hoy)
```

```python
#creo el Dataset
horas = list()
temp = list()
v_viento = list()

for i in range(0, len(tiempo_hoy), 7):
    horas.append(tiempo_hoy[i])
    temp.append(tiempo_hoy[i+2])
    v_viento.append(tiempo_hoy[i+5])

df = pd.DataFrame({'Horas': horas, 'Temperatura': temp, 'V_viento(km_h)':v_viento})

df
```
