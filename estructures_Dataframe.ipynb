```python
#recoger todo el HTML de la página
import requests

URL = "https://www.bolsamadrid.es/esp/aspx/Mercados/Precios.aspx?indice=ESI100000000"
page = requests.get(URL)

print(page.text)
```

```python
#creación de objeto Beautiful soup
from bs4 import BeautifulSoup

URL = "https://www.bolsamadrid.es/esp/aspx/Mercados/Precios.aspx?indice=ESI100000000"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")


results = soup.find(id="ctl00_Contenido_tblAcciones") 

print(results.prettify())
```

```python
#Ver los elementos de la página
print(set([text.parent.name for text in soup.find_all(text=True)]))
```

```python
elements = results.find_all("td") 
elements
```

```python
for element in elements:
    print(element, end="\n"*2)
```

```python
#Creo un df con la info
tables = soup.findChildren("table", { "id" :"ctl00_Contenido_tblAcciones" })

my_table = tables[0]

rows = my_table.findChildren(['tbody', 'tr'])

names = my_table.findChildren(['tbody', 'th'])

columns_name = []

for name in names:
    columns_name.append(name.text)

df_bolsa = pd.DataFrame(columns=columns_name)

for row in rows:
    cols = []
    for x in row.findAll("td"):  
      cols.append(x.text)
    if len(cols)!=0:
      df_bolsa.loc[len(df_bolsa.index)] = cols

df_bolsa
```

Luego tomo la [página de Quotes](http://quotes.toscrape.com) para hacer el mismo trabajo
[Fuente](https://github.com/pepovinueza/Web_Scraping/blob/main/Sprint%2010.%20Web%20Scrapping.ipynb)

```python
URL2 = 'http://quotes.toscrape.com'
page2 = requests.get(URL2)

print(page2.text)
```

```python
soup2 = BeautifulSoup(page2.content, "html.parser")

elements = soup2.find_all("div", class_="quote")
elements
```

```python
df_quotes = pd.DataFrame()
for element in elements:
    quote = element.find("span", class_="text")
    author = element.find("small", class_="author")
    tags = element.find("a", class_="tag")
    df_quotes = df_quotes.append({'Frase' : quote.text.strip(),
                    'Autor' : author.text.strip(),
                     'Tag': tags.text.strip()},
                    ignore_index = True)
    
df_quotes
```

```python
driver.get(URL)
```

```python
columns_name2 = []

for name in driver.find_elements(By.TAG_NAME, 'th'):
    columns_name2.append(name.text)

columns_name2 = columns_name2[-9:]

columns_name2
```

```python
df_bolsa2 = pd.DataFrame(columns=columns_name2)

table = driver.find_element(By.ID, 'ctl00_Contenido_tblAcciones') 

df = []

for i in table.find_elements(By.TAG_NAME, 'td'): 
    df.append(i.text) 
    row = [df[x:x+len(columns_name2)] for x in range(0, len(df), len(columns_name2))] 
    
df_bolsa2 = pd.DataFrame(row, columns=columns_name2) 
df_bolsa2
```

```python
sentences = driver.find_elements(By.CLASS_NAME, "text")
authors = driver.find_elements(By.CLASS_NAME, "author")

sentences_list = []

for sentence in sentences:
    sentence = sentence.text.strip()
    sentences_list.append(sentence)
    
authors_list = []

for author in authors:
    author = author.text.strip()
    authors_list.append(author)
    
    
df_sentences = pd.DataFrame()

df_sentences['Frase'] = sentences_list
df_sentences['Autor'] = authors_list

df_sentences
```
